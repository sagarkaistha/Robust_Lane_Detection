```python
import matplotlib.pyplot as plt
import numpy as np

# ============================================================
# DATA: All three models
# ============================================================
scenarios = ['Normal', 'Crowd', 'Highlight', 'Shadow', 'No Line', 'Arrow', 'Curve', 'Night']

unet_iou =  [0.614, 0.469, 0.313, 0.328, 0.148, 0.543, 0.398, 0.107]
hog_iou =   [0.571, 0.412, 0.530, 0.517, 0.437, 0.580, 0.545, 0.350]
lcnn_iou =  [0.32,  0.20,  0.17,  0.17,  0.16,  0.17,  0.27,  0.13]

unet_f1 =   [0.756, 0.602, 0.466, 0.483, 0.239, 0.698, 0.552, 0.157]
hog_f1 =    [0.727, 0.583, 0.693, 0.681, 0.608, 0.734, 0.706, 0.518]
lcnn_f1 =   [0.48,  0.32,  0.28,  0.28,  0.27,  0.28,  0.41,  0.23]

# ============================================================
# PLOT 1: Training & Validation Loss (U-Net)
# ============================================================
train_losses = [0.606, 0.469, 0.440, 0.424, 0.415, 0.409, 0.404]
val_losses =   [0.564, 0.541, 0.540, 0.527, 0.528, 0.530, 0.520]

plt.figure(figsize=(8, 5))
plt.plot(range(1, 8), train_losses, 'b-o', label='Train Loss', linewidth=2)
plt.plot(range(1, 8), val_losses, 'r-o', label='Val Loss', linewidth=2)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss (BCE + Dice)', fontsize=12)
plt.title('U-Net: Training & Validation Loss (88K images)', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plot_training_loss.png', dpi=150)
plt.show()

# ============================================================
# PLOT 2: IoU Bar Chart — All 3 Models
# ============================================================
x = np.arange(len(scenarios))
width = 0.25

fig, ax = plt.subplots(figsize=(14, 6))
bars1 = ax.bar(x - width, unet_iou, width, label='U-Net (ours)', color='#3498db', edgecolor='black', linewidth=0.5)
bars2 = ax.bar(x, hog_iou, width, label='HOG + LR', color='#e74c3c', edgecolor='black', linewidth=0.5)
bars3 = ax.bar(x + width, lcnn_iou, width, label='Lightweight CNN', color='#2ecc71', edgecolor='black', linewidth=0.5)

ax.set_ylabel('IoU', fontsize=12)
ax.set_title('Per-Scenario IoU: U-Net vs HOG+LR vs Lightweight CNN', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(scenarios, rotation=45, ha='right', fontsize=10)
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)

for bars in [bars1, bars2, bars3]:
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.008,
                f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=7)

plt.tight_layout()
plt.savefig('plot_iou_comparison.png', dpi=150)
plt.show()

# ============================================================
# PLOT 3: F1 Bar Chart — All 3 Models
# ============================================================
fig, ax = plt.subplots(figsize=(14, 6))
bars1 = ax.bar(x - width, unet_f1, width, label='U-Net (ours)', color='#3498db', edgecolor='black', linewidth=0.5)
bars2 = ax.bar(x, hog_f1, width, label='HOG + LR', color='#e74c3c', edgecolor='black', linewidth=0.5)
bars3 = ax.bar(x + width, lcnn_f1, width, label='Lightweight CNN', color='#2ecc71', edgecolor='black', linewidth=0.5)

ax.set_ylabel('F1 Score', fontsize=12)
ax.set_title('Per-Scenario F1: U-Net vs HOG+LR vs Lightweight CNN', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(scenarios, rotation=45, ha='right', fontsize=10)
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)

for bars in [bars1, bars2, bars3]:
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.008,
                f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=7)

plt.tight_layout()
plt.savefig('plot_f1_comparison.png', dpi=150)
plt.show()

# ============================================================
# PLOT 4: Radar Chart — All 3 Models (IoU)
# ============================================================
angles = np.linspace(0, 2 * np.pi, len(scenarios), endpoint=False).tolist()
unet_r = unet_iou + [unet_iou[0]]
hog_r = hog_iou + [hog_iou[0]]
lcnn_r = lcnn_iou + [lcnn_iou[0]]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(9, 9), subplot_kw=dict(polar=True))
ax.plot(angles, unet_r, 'b-o', linewidth=2, label='U-Net (ours)', markersize=6)
ax.fill(angles, unet_r, alpha=0.1, color='blue')
ax.plot(angles, hog_r, 'r-s', linewidth=2, label='HOG + LR', markersize=6)
ax.fill(angles, hog_r, alpha=0.1, color='red')
ax.plot(angles, lcnn_r, 'g-^', linewidth=2, label='Lightweight CNN', markersize=6)
ax.fill(angles, lcnn_r, alpha=0.1, color='green')

ax.set_xticks(angles[:-1])
ax.set_xticklabels(scenarios, fontsize=10)
ax.set_ylim(0, 0.7)
ax.set_title('IoU Comparison Across Scenarios', fontsize=14, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.35, 1.1), fontsize=11)
plt.tight_layout()
plt.savefig('plot_radar.png', dpi=150)
plt.show()

# ============================================================
# PLOT 5: Radar Chart — All 3 Models (F1)
# ============================================================
unet_f1_r = unet_f1 + [unet_f1[0]]
hog_f1_r = hog_f1 + [hog_f1[0]]
lcnn_f1_r = lcnn_f1 + [lcnn_f1[0]]

fig, ax = plt.subplots(figsize=(9, 9), subplot_kw=dict(polar=True))
ax.plot(angles, unet_f1_r, 'b-o', linewidth=2, label='U-Net (ours)', markersize=6)
ax.fill(angles, unet_f1_r, alpha=0.1, color='blue')
ax.plot(angles, hog_f1_r, 'r-s', linewidth=2, label='HOG + LR', markersize=6)
ax.fill(angles, hog_f1_r, alpha=0.1, color='red')
ax.plot(angles, lcnn_f1_r, 'g-^', linewidth=2, label='Lightweight CNN', markersize=6)
ax.fill(angles, lcnn_f1_r, alpha=0.1, color='green')

ax.set_xticks(angles[:-1])
ax.set_xticklabels(scenarios, fontsize=10)
ax.set_ylim(0, 0.8)
ax.set_title('F1 Score Comparison Across Scenarios', fontsize=14, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.35, 1.1), fontsize=11)
plt.tight_layout()
plt.savefig('plot_radar_f1.png', dpi=150)
plt.show()

# ============================================================
# PLOT 6: Mean IoU & F1 Summary
# ============================================================
models = ['U-Net\n(ours)', 'HOG + LR', 'Lightweight\nCNN']
mean_iou = [np.mean(unet_iou), np.mean(hog_iou), np.mean(lcnn_iou)]
mean_f1 = [np.mean(unet_f1), np.mean(hog_f1), np.mean(lcnn_f1)]

x = np.arange(len(models))
width = 0.3

fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, mean_iou, width, label='Mean IoU', color='#3498db', edgecolor='black', linewidth=0.5)
bars2 = ax.bar(x + width/2, mean_f1, width, label='Mean F1', color='#e67e22', edgecolor='black', linewidth=0.5)

for bar in bars1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.008,
            f'{bar.get_height():.3f}', ha='center', va='bottom', fontsize=10, fontweight='bold')
for bar in bars2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.008,
            f'{bar.get_height():.3f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

ax.set_ylabel('Score', fontsize=12)
ax.set_title('Mean IoU & F1 Across All Scenarios', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=11)
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('plot_mean_comparison.png', dpi=150)
plt.show()

# ============================================================
# PLOT 7: Training Progression (10k vs 50k vs 88k)
# ============================================================
datasets = ['10K\n(no aug)', '50K\n(+aug, +dropout)', '88K\n(full dataset)']
normal_prog = [0.45, 0.588, 0.614]
crowd_prog = [0.05, 0.097, 0.469]
night_prog = [0.00, 0.003, 0.107]

x = np.arange(len(datasets))
width = 0.25

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(x - width, normal_prog, width, label='Normal', color='#2ecc71', edgecolor='black', linewidth=0.5)
ax.bar(x, crowd_prog, width, label='Crowd', color='#3498db', edgecolor='black', linewidth=0.5)
ax.bar(x + width, night_prog, width, label='Night', color='#9b59b6', edgecolor='black', linewidth=0.5)

ax.set_ylabel('IoU', fontsize=12)
ax.set_title('Impact of Data Size & Regularisation on IoU', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(datasets, fontsize=10)
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('plot_progression.png', dpi=150)
plt.show()

# ============================================================
# PLOT 8: Parameter Count vs Mean IoU
# ============================================================
model_names = ['Lightweight CNN\n(93K params)', 'HOG + LR\n(N/A)', 'U-Net\n(31M params)']
params = [93377, 50000, 31043521]  # HOG params approximate
mean_ious = [np.mean(lcnn_iou), np.mean(hog_iou), np.mean(unet_iou)]
colors_scatter = ['#2ecc71', '#e74c3c', '#3498db']

fig, ax = plt.subplots(figsize=(8, 5))
for i in range(3):
    ax.scatter(params[i], mean_ious[i], s=200, c=colors_scatter[i], edgecolors='black', linewidth=1, zorder=5)
    ax.annotate(model_names[i], (params[i], mean_ious[i]),
                textcoords="offset points", xytext=(10, 10), fontsize=9)

ax.set_xscale('log')
ax.set_xlabel('Parameters (log scale)', fontsize=12)
ax.set_ylabel('Mean IoU', fontsize=12)
ax.set_title('Model Complexity vs Performance', fontsize=14)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plot_params_vs_iou.png', dpi=150)
plt.show()
```
